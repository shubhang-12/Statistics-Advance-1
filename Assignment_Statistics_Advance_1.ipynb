```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
  """
  Calculates the probability distribution function (PDF) of a Bernoulli distribution.

  Args:
    x: The value of the random variable (0 or 1).
    p: The probability of success (value 1).

  Returns:
    The probability of observing the given value x.
  """
  if x == 0:
    return 1 - p
  elif x == 1:
    return p
  else:
    return 0

# The following lines were incorrectly indented, causing the error.
# They should be aligned with the function call above.
probability = bernoulli_pdf(1, 0.6)  # Probability of success (x=1) with p=0.6
print(probability)  # Output: 0.6
```

```python
#Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(x, low, high):
  """
  Calculates the cumulative distribution function (CDF) of a discrete uniform distribution.

  Args:
    x: The value at which to evaluate the CDF.
    low: The lower bound of the distribution.
    high: The upper bound of the distribution.

  Returns:
    The CDF value at x.
  """
  if x < low:
    return 0
  elif x >= high:
    return 1
  else:
    return (x - low + 1) / (high - low + 1)

# Define the parameters of the distribution
low = 1
high = 6

# Generate a range of x values
x_values = np.arange(low - 1, high + 2)

# Calculate the CDF values for each x
cdf_values = [discrete_uniform_cdf(x, low, high) for x in x_values]

# Plot the CDF
plt.step(x_values, cdf_values)
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
plt.grid(True)
plt.show()
```

```python
#Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_zscores(data):
    """Calculates Z-scores from a dataset and plots them.

    Args:
        data (list or numpy array): The dataset.
    """

    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot the Z-scores using Seaborn
    sns.histplot(z_scores, kde=True)  # kde=True adds a kernel density estimate curve
    plt.title("Distribution of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.show()

    return z_scores

# Example usage:
# Changed 'your_data' to 'data' to use the existing variable
z_scores = calculate_and_plot_zscores(data)
```

```python
#Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_standard_normal_distribution():
  """Calculates and plots the standard normal distribution.

  The standard normal distribution has a mean of 0 and a standard deviation of 1.
  """
  # Generate x-values for the plot
  x = np.linspace(-4, 4, 100)  # Create 100 points between -4 and 4

  # Calculate the probability density function (PDF) for each x-value
  pdf = stats.norm.pdf(x, loc=0, scale=1)  # loc is mean, scale is standard deviation

  # Plot the PDF
  plt.plot(x, pdf)
  plt.title("Standard Normal Distribution")
  plt.xlabel("x")
  plt.ylabel("Probability Density")
  plt.show()

# Call the function to create the plot
plot_standard_normal_distribution()
```

```python
#Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import scipy.stats as stats

def calculate_zscore(data_point, data_set):
    """Calculates the Z-score for a given data point.

    Args:
        data_point: The data point to calculate the Z-score for.
        data_set: The data set to use for the calculation.

    Returns:
        The Z-score for the data point.
    """
    mean = np.mean(data_set)
    std = np.std(data_set)
    z_score = (data_point - mean) / std
    return z_score

# Example usage
data_point = 75
data_set = [60, 65, 70, 75, 80, 85, 90]
z_score = calculate_zscore(data_point, data_set)

# Compare to standard normal distribution
p_value = stats.norm.cdf(z_score)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats

# Sample data
sample_data = [60, 65, 70, 75, 80, 85, 90]

# Population parameters (hypothesized)
population_mean = 70
population_std = 10

# 1. State the hypotheses:
# Null hypothesis (H0): The sample mean is equal to the population mean (no difference).
# Alternative hypothesis (H1): The sample mean is not equal to the population mean (there is a difference).

# 2. Set the significance level (alpha):
alpha = 0.05  # 5% significance level

# 3. Calculate the sample statistics:
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# 4. Calculate the Z-statistic:
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# 5. Determine the p-value:
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))  # Two-tailed test

# 6. Make a decision:
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("There is evidence to suggest that the sample mean is different from the population mean.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is not enough evidence to suggest that the sample mean is different from the population mean.")

# Print the results:
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
#Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats
data = [60, 65, 70, 75, 80, 85, 90]  # Replace with your actual dataset
confidence_level = 0.95  # 95% confidence level
degrees_of_freedom = len(data) - 1  # Degrees of freedom
sample_mean = np.mean(data)
sample_standard_error = stats.sem(data)  # Standard error of the mean

confidence_interval = stats.t.interval(
    confidence_level, degrees_of_freedom, sample_mean, sample_standard_error
)
print(f"The {confidence_level*100}% confidence interval for the mean is: {confidence_interval}")
```

```python
#Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Example dataset
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Or, load your dataset from a file:
# df = pd.read_csv('your_dataset.csv')
# Calculate the mean and standard deviation
mean = df['Values'].mean()
std = df['Values'].std()

# Apply Z-score normalization
df['Z-score Normalized'] = (df['Values'] - mean) / std
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot the original data
sns.histplot(df['Values'], kde=True, ax=ax1)
ax1.set_title('Original Data Distribution')

# Plot the Z-score normalized data
sns.histplot(df['Z-score Normalized'], kde=True, ax=ax2)
ax2.set_title('Z-score Normalized Data Distribution')

plt.show()
```

```python
#Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):
  """Calculates the skewness and kurtosis of a dataset.

  Args:
    data: A NumPy array or a list of data values.

  Returns:
    A tuple containing the skewness and kurtosis of the data.
  """
  skewness = stats.skew(data)
  kurtosis = stats.kurtosis(data)
  return skewness, kurtosis

# Generate data from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate skewness and kurtosis
skewness, kurtosis = calculate_skewness_kurtosis(data)

# Print the results
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```
